```
fig, ax = plt.subplots(figsize=(10,8))
plt.suptitle('')
df.boxplot(column=['NO2'], by='City', ax=ax)
```

```
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", palette="pastel")


sns.boxplot(x="City", y="NO2",
            hue="Year", palette=["b", "r"],
            data=df)
sns.despine(offset=10, trim=True)
```

# Seaborn violinplot

```
import seaborn as sns
sns.set_theme(style="whitegrid")
```

```
ax = sns.violinplot(x=df["NO2"])
```

```
ax = sns.violinplot(x="City", y="NO2", data=df)
```

```
ax = sns.violinplot(x="Year", y="NO2", data=df)
```

```
ax = sns.violinplot(x="Year", y="NO2", hue="City",
                    data=df, palette="muted")
```

```
ax = sns.violinplot(x="City", y="NO2", hue="Year",
                    data=df, palette="muted")
```

```
ax = sns.violinplot(x="City", y="NO2", hue="Year",
                    data=df, palette="muted", split=True)
```

```
ax = sns.violinplot(x="City", y="NO2", hue="Year",
                    data=df, palette="muted", split=True, scale='count')
```

```
ax = sns.violinplot(x="City", y="NO2", hue="Year",
                    data=df, palette="muted", split=True, scale='count', inner="quartile")
```

# Stacked histogram

```
sns.histplot(
    df,
    x="City", hue="Year",
    multiple="stack",
    palette="light:b_r",
    edgecolor=".3",
    linewidth=.5,
)
ax.xaxis.set_major_formatter(mpl.ticker.ScalarFormatter())
```

```
g = sns.catplot(
    data=df, kind="bar",
    x="City", y="NO2", hue="Year",
    ci="sd", palette="dark", alpha=.6, height=6
)
g.despine(left=True)
g.set_axis_labels("", "NO2")
g.legend.set_title("Year")
```

# Scatterplot with categorical variables

```
ax = sns.swarmplot(data=df, x="City", y="NO2", hue="Year")
ax.set(ylabel="")
```

```
sns.set_theme(style="whitegrid", palette="muted")
ax = sns.swarmplot(data=df, x="Year", y="NO2", hue="City")
ax.set(ylabel="")
```

# Line plot

```
sns.set_theme(style="whitegrid", palette="muted")
sns.lineplot(x="City", y="NO2",
             hue="Year",
             data=df)
```

```
sns.set_theme(style="whitegrid", palette="muted")
sns.lineplot(x="City", y="NO2",
             hue="Year", style="Year",
             data=df)
```

# Pie chart

```
df_pie_Year = df.groupby('Year')['NO2'].nunique()

df_pie_Year.plot.pie(y='City', figsize=(5, 5),subplots=True)

df_pie_Year
```

```
df_pie_City = df.groupby('City')['NO2'].nunique()

df_pie_City.plot.pie(y='City', figsize=(5, 5))

df_pie_City
```

```
explode_Madrid = (0, .5, 0, 0, 0, 0)
df_pie_City.plot.pie(y='City', figsize=(5, 5), explode = explode_Madrid)
```

```
explode_New_Delhi = (0, 0, 0.1, 0, 0, 0)
df_pie_City.plot.pie(y='City', figsize=(5, 5), explode = explode_New_Delhi)

df_pie_City
```
