```python
import pandas as pd   # Nested loop

# Load the DataFrame
df = pd.read_csv("NO2.csv")  # Adjust this as necessary to match your file path and DataFrame

# Get unique years and cities from the DataFrame
unique_years = df['Year'].unique()
unique_cities = df['City'].unique()

# Outer loop over years
for year in unique_years:
    print(f"\nYear: {year}")  # Print the year for clarity in output
    # Inner loop over cities
    for city in unique_cities:
        # Filter df for the current year and city
        filtered_df = df[(df['Year'] == year) & (df['City'] == city)]

        # Calculate mean NO2 if there are entries for the filter
        if not filtered_df.empty:
            mean_no2 = filtered_df['NO2'].mean()
            print(f"Average NO2 level for {city} in {year}: {mean_no2:.6f}")
        else:
            print(f"No data for {city} in {year}.")
```

```python
import pandas as pd

# Example: Reading and writing CSV files
data = {"lat": [34.0522, 40.7128, 37.7749], "lon": [-118.2437, -74.0060, -122.4194], "city": ["Delhi", "Prayagraj", "Boston"]}
df = pd.DataFrame(data)
df.to_csv("cities.csv", index=False)
df_read = pd.read_csv("cities.csv")
print(df_read.head())
```

```python
# Example: Visualization of tabular GIS data using Matplotlib
import matplotlib.pyplot as plt

plt.scatter(df_read["lon"], df_read["lat"])
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City Locations")
plt.show()
```

```python
plt.scatter(df_read["lon"], df_read["lat"])
for i, txt in enumerate(df_read["city"]):
    plt.annotate(txt, (df_read["lon"][i], df_read["lat"][i]), fontsize=9, ha='right')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City Locations with Labels")
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_read["lon"], df_read["lat"], c='red')  # All points are blue
for i, txt in enumerate(df_read["city"]):
    plt.annotate(txt, (df_read["lon"][i], df_read["lat"][i]), fontsize=9, ha='right', color='red')  # Annotation in red
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("City Locations with Labels")
plt.show()
```

```python
import geopandas as gpd
from shapely.geometry import Point

# Example: Creating and manipulating GeoDataFrames
geometry = [Point(xy) for xy in zip(df_read["lon"], df_read["lat"])]
gdf = gpd.GeoDataFrame(df_read, geometry=geometry)
print(gdf.head())

# Example: Plotting spatial data using Matplotlib
gdf.plot()
plt.title("Geospatial City Locations")
plt.show()
```

```python
import folium
import random

def generate_random_coordinates(lat, lon, num_coords):
    """Generate random coordinates around a central point (latitude, longitude)."""
    return [(lat + random.uniform(-1, 1), lon + random.uniform(-1, 1)) for _ in range(num_coords)]

# Create a map centered around the geographical center of the contiguous US
map_us = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

# Generate random locations
random_locations = generate_random_coordinates(37.0902, -95.7129, 10)  # Generates 10 random points

# Add a marker for the center location
folium.Marker([37.0902, -95.7129], popup='Center of the US').add_to(map_us)

# Loop through the list of random coordinates and add them as markers to the map
for lat, lon in random_locations:
    folium.Marker([lat, lon], popup=f'Random Point: {lat:.2f}, {lon:.2f}').add_to(map_us)

# Display the map
map_us
```

```python
import folium
import random
import csv

def generate_random_coordinates(lat, lon, num_coords):
    """Generate random coordinates around a central point (latitude, longitude)."""
    return [(lat + random.uniform(-1, 1), lon + random.uniform(-1, 1)) for _ in range(num_coords)]

# Create a map centered around the geographical center of the contiguous US
map_us = folium.Map(location=[37.0902, -95.7129], zoom_start=5)

# Generate random locations
random_locations = generate_random_coordinates(37.0902, -95.7129, 10)  # Generates 10 random points

# Add a marker for the center location
folium.Marker([37.0902, -95.7129], popup='Center of the US').add_to(map_us)

# Loop through the list of random coordinates and add them as markers to the map
for lat, lon in random_locations:
    folium.Marker([lat, lon], popup=f'Random Point: {lat:.2f}, {lon:.2f}').add_to(map_us)

# Save the map as an HTML file
map_us.save('map.html')

# Write coordinates to a CSV file
with open('coordinates.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Latitude', 'Longitude'])  # Write header
    for lat, lon in random_locations:
        writer.writerow([lat, lon])

# Print out the coordinates
print("Coordinates saved to 'coordinates.csv':")
for lat, lon in random_locations:
    print(f"Latitude: {lat:.4f}, Longitude: {lon:.4f}")
# Display the map
map_us
```
